```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = datasets.load_iris()
```

```python
X_reduced = PCA(n_components=3).fit_transform(iris.data)
Y = iris.target

fig = plt.figure(1, figsize=(8, 8))
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, cmap=plt.cm.rainbow_r)

ax.set_title("Three PCA directions", fontsize = 20)
ax.set_xlabel("1st Eigenvector", fontsize = 15)
ax.set_ylabel("2nd Eigenvector", fontsize = 15)
ax.set_zlabel("3rd Eigenvector", fontsize = 15)

plt.show()
```
